```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
sns.set()

df = pd.read_csv("/content/amazon_clean.csv")

print("Shape:", df.shape)
print("\nKolom yang tersedia:")
print(df.columns.tolist())

df.head()
```

```python
# Info tipe data
df.info()

# Statistik numerik
df.describe()

# Statistik untuk object/categorical (optional)
df.describe(include="object")
```

```python
prioritas = [
    "discounted_price",
    "actual_price",
    "discount_percentage",
    "rating",
    "rating_count"
]

num_cols = [c for c in prioritas if c in df.columns]
print("Kolom numerik terpilih:", num_cols)
```

```python
if num_cols:
    df[num_cols].hist(bins=30, figsize=(14, 8))
    plt.suptitle("Histogram Kolom Numerik", y=1.02)
    plt.tight_layout()
    plt.savefig("assets/images/hist_all_numeric.png", bbox_inches="tight")
    plt.show()
else:
    print("Tidak ada kolom numerik yang ditemukan.")
```

```python
if num_cols:
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df[num_cols])
    plt.xticks(rotation=45)
    plt.title("Boxplot Kolom Numerik")
    plt.tight_layout()
    plt.savefig("assets/images/boxplot_numeric.png", bbox_inches="tight")
    plt.show()
```

```python
# Kandidat kolom kategorikal utama
candidate_cat_cols = ["main_category", "category"]

cat_cols = [c for c in candidate_cat_cols if c in df.columns]
print("Kolom kategori yang dipakai:", cat_cols)

for col in cat_cols:
    plt.figure(figsize=(10, 6))
    df[col].value_counts().head(15).plot(kind="bar")
    plt.title(f"Top 15 {col} by Count")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=75)
    plt.tight_layout()
    plt.savefig(f"assets/images/top15_{col}.png", bbox_inches="tight")
    plt.show()
```

```python
if len(num_cols) >= 2:
    plt.figure(figsize=(8, 6))
    corr = df[num_cols].corr()
    sns.heatmap(corr, annot=True, fmt=".2f", cmap="Blues")
    plt.title("Correlation Matrix (Kolom Numerik)")
    plt.tight_layout()
    plt.savefig("assets/images/heatmap_correlation.png", bbox_inches="tight")
    plt.show()
else:
    print("Tidak cukup kolom numerik untuk membuat heatmap korelasi.")
```

```python
pairs = [
    ("discounted_price", "rating"),
    ("discount_percentage", "rating_count"),
    ("actual_price", "discounted_price"),
]

for x_col, y_col in pairs:
    if x_col in df.columns and y_col in df.columns:

        plt.figure(figsize=(8, 5))
        sns.scatterplot(data=df, x=x_col, y=y_col, alpha=0.3)
        plt.title(f"Scatter Plot: {x_col} vs {y_col}")
        plt.tight_layout()

        # Buat nama file AMAN
        safe_name = f"{x_col}_vs_{y_col}".replace(" ", "_")

        # Simpan file
        plt.savefig(f"assets/images/scatter_{safe_name}.png", bbox_inches="tight")

        # Tampilkan plot
        plt.show()

    else:
        print(f"Skip scatter {x_col} vs {y_col} (kolom tidak lengkap).")
```

```python
if "review_content" in df.columns:
    df["review_length"] = df["review_content"].astype(str).str.len()
    plt.figure(figsize=(8, 5))
    sns.histplot(df["review_length"], bins=30)
    plt.title("Distribusi Panjang Review")
    plt.tight_layout()
    plt.savefig("assets/images/hist_review_length.png", bbox_inches="tight")
    plt.show()
```
